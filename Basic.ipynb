# Dicision Tree

```python
from sklearn import tree
x = [[0,0], [1,1]]
y = [0,1]
```

```python
# 다양한 옵션. 알고리즘에 따라 데이터 분류

clf = tree.DecisionTreeClassifier()
clf
```

```python
# 학습
clf = clf.fit(x,y)
clf
```

```python
# 1이라는 값을 예측 : 정답값이 0과 1로만 학습이 되어서 1이라는 결과값
clf.predict([[2.,2.]])
```

```python
# 0., 1. 이라는 값을 예측 : 비율로 출력
clf.predict_proba([[2.,2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
x, y = load_iris(return_X_y = True)
x, y
# y ; 품종 categorical, dicisiontree 알고리즘은 숫자만 인식할 수 있어서 0 1 2 - 분류값
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
t = tree.plot_tree(clf.fit(x,y), filled=True)
```
